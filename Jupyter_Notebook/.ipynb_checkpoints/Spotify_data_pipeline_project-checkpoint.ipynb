```python

# This is used for the authorization purpose
sp = spotipy.Spotify(client_credentials_manager = client_credetials_manager)
```

```python
# Copy the URL from the Spotify webiste
playlist_link = 'https://open.spotify.com/playlist/1bvcVgSFFKobfJOWgsUrEe'
```

```python
# Fetch the last id from the URL
playlist_URI = playlist_link.split("/")[-1]
```

```python
# make the object 'data' that fetch the raw data from the spotify API
data = sp.playlist_items(playlist_URI)
```

```python
# create album list from the raw data
album_list = []
for row in data['items']:
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    album_release_date = row['track']['album']['release_date']
    album_total_tracks = row['track']['album']['total_tracks']
    album_url = row['track']['album']['external_urls']['spotify']
    album_dic = {'album_id' : album_id, 'album_name' : album_name, 'album_release_date' : album_release_date, 'album_total_tracks' : album_total_tracks, 'album_url' : album_url}
    album_list.append(album_dic)
```

```python
# create the artist_list that stores all artist data
artist_list = []
for row in data['items']:
    for key, value in row.items():
        if key == 'track':
            for artist in value['artists']:
                artist_dic = {'artist_id': artist['id'], 'artist_name' : artist['name'], 'external_url' : artist['href']}
                artist_list.append(artist_dic)
                
        
```

```python
# print artist list
artist_list
```

```python
data['items'][0]['track']
```

```python
# create the song list
song_list = []
for row in data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularly = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_element = {
        'song_id' : song_id, 'song_name' : song_name, 'song_duration' : song_duration, 'song_url' : song_url, 'song_popularly' : song_popularly, 
        'song_added' : song_added, 'album_id' : album_id, 'artist_id' : artist_id
    }
    song_list.append(song_element)
    
```

```python
# print song list
song_list
```

```python
# convert album list into dataframe
album_df = pd.DataFrame.from_dict(album_list)
```

```python
# convert artist list into dataframe
artist_df = pd.DataFrame.from_dict(artist_list)
```

```python
# convert song list into dataframe
song_df = pd.DataFrame.from_dict(song_list)
```

```python
song_df
```

```python
# drop duplicates from the album df (cleaning process)
album_df = album_df.drop_duplicates(subset = ['album_id'])
```

```python
# drop duplicates from the artist df (cleaning process)
artist_df = artist_df.drop_duplicates(subset = ['artist_id'])
```

```python
# drop duplicates from the songs df (cleaning process)
song_df = song_df.drop_duplicates(subset = ['song_id'])
```

```python
# convert 'album_release_date' into the DateTime object
album_df['album_release_date'] = pd.to_datetime(album_df['album_release_date'])
```

```python
# convert 'song_added' into the DateTime object
song_df['song_added'] = pd.to_datetime(song_df['song_added'])
```
